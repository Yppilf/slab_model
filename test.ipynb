```python
import pandas as pd
from matplotlib.pyplot import figure,show,savefig
import matplotlib.ticker as mticker
```

```python
df = pd.read_csv("models/slab_model_H2O_400K_5-25um_1e+18cm-2.out")
wavelengths = df["convWavelength[microns]"].to_numpy()
print(df)
print(wavelengths[0], wavelengths[-1])
step_sizes = [j - i for i, j in zip(wavelengths[:-1], wavelengths[1:])]
```

```python
fig = figure()
frame = fig.add_subplot()
frame.plot(wavelengths[:-1], step_sizes)
frame.set_title("Step size evolution of wavelength")
frame.set_xlabel("Convolved wavelength [$\mu$m]")
frame.set_ylabel("Step size in wavelength")
frame.yaxis.set_major_formatter(mticker.ScalarFormatter(useMathText=False))
frame.yaxis.get_major_formatter().set_useOffset(False)
frame.xaxis.get_major_formatter().set_useOffset(False)
# frame.set_xticklabels(frame.get_xticklabels(), rotation=45)
savefig("Stepsize_evolution.pdf")
```
